```python
N = int(input())
```

```python
import os
from collections import deque

ans = 0

def dfs(s):
    global ans
    q = deque([])   # スタック
    q.append(s)
    
    while len(q) > 0:
        print(q)
        p = q.pop()         # スタック取りだし（末尾）
        print(p)
        print(q)
        if ('7' in p) and ('5' in p) and ('3' in p):
            ans += 1
        p_child_1 = p + '7'
        p_child_2 = p + '5'
        p_child_3 = p + '3'
        if int(p_child_1) < N:
            q.append(p_child_1)   # スタック追加
        if int(p_child_2) < N:
            q.append(p_child_2)   # スタック追加
        if int(p_child_3) < N:
            q.append(p_child_3)   # スタック追加
    return ans

print(dfs('0'))
```
