```python
H = list(map(int, input().split()))
```

```python
x = H[0]
a = 0
max = 0
count = 0
```

```python
print(len(H))
```

```python
while a < len(H):
    for i in range(a,len(H)):
        print(x)
        if x >= H[i]:
            count = count + 1
            x = H[i]
        
        if x < H[i]:
            x = H[i]
            a = i
            
        if max < count:
            max = count
            count = 0
```
